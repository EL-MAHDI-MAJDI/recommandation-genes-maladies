# Phase 1 — Compréhension Théorique et Cadrage du Projet

## Contexte

Ce projet s'inscrit dans le domaine du **Graph Mining** et vise à exploiter les techniques de **prédiction de liens (Link Prediction)** pour découvrir de nouvelles associations potentielles entre des gènes et des maladies.

L'objectif est de construire un modèle capable d'analyser un réseau biologique connu et de prédire des relations qui n'ont pas encore été observées ou validées scientifiquement.

Cette approche est particulièrement intéressante en bioinformatique, car les bases de données actuelles ne contiennent qu'une partie des interactions réellement existantes entre les gènes et les maladies.

# 1. Qu'est-ce que la prédiction de liens (Link Prediction) ?

La prédiction de liens est un problème fondamental du Graph Mining.

Elle consiste à estimer la probabilité qu'une connexion (arête) existe entre deux nœuds qui ne sont actuellement pas reliés dans un graphe.

À partir de la structure existante du réseau, l'objectif est de découvrir :

- des liens manquants ;
- des liens futurs ;
- des relations encore inconnues.

Les méthodes de prédiction utilisent généralement :

- la topologie du graphe ;
- les voisins communs ;
- les mesures de similarité ;
- les distances entre nœuds ;
- les marches aléatoires (Random Walks) ;
- les représentations apprises par Machine Learning.

La prédiction de liens est largement utilisée dans :

- les réseaux sociaux ;
- les systèmes de recommandation ;
- les réseaux biologiques ;
- les réseaux de connaissances.

# 2. Application au problème Gène–Maladie

Dans ce projet, nous utilisons un graphe biparti composé de deux types de nœuds :

- les gènes ;
- les maladies.

Une arête relie un gène à une maladie lorsqu'une association a déjà été validée dans la littérature scientifique.

Le graphe peut être représenté comme suit :

Gène ←→ Maladie

Cependant, toutes les associations réelles ne sont pas encore connues.

Le problème devient donc :

> Peut-on prédire quelles paires (gène, maladie) non connectées aujourd'hui ont une forte probabilité d'être associées dans le futur ?

Cette formulation est largement utilisée dans la littérature scientifique pour la découverte de gènes impliqués dans des maladies humaines.

Ainsi, la prédiction de liens permet de générer des hypothèses biologiques qui pourront être validées ultérieurement par des chercheurs.

# 3. État de l'Art

De nombreuses études récentes ont montré l'efficacité des méthodes de Graph Mining pour la prédiction des associations gène-maladie.

## Réseaux de neurones de graphes (GNN)

Les Graph Neural Networks (GNN) permettent d'exploiter directement la structure du graphe.

Plusieurs travaux ont obtenu d'excellentes performances pour la prédiction des associations gène-maladie.

Dans certaines études, la majorité des associations proposées par le modèle ont ensuite été confirmées comme plausibles dans la littérature scientifique.

## Variational Graph AutoEncoder (VGAE)

Le VGAE est une approche non supervisée qui apprend une représentation latente des nœuds du graphe.

Ces représentations peuvent ensuite être utilisées pour prédire de nouvelles connexions entre gènes et maladies.

## Graphes hétérogènes multi-sources

Des travaux plus récents combinent plusieurs réseaux biologiques :

- graphe gène-gène ;
- graphe maladie-maladie ;
- graphe gène-maladie.

L'intégration de ces différentes sources d'information améliore généralement les performances de prédiction.

Ces résultats montrent que la prédiction de liens constitue une approche prometteuse pour aider à la découverte de nouveaux biomarqueurs et de nouvelles cibles thérapeutiques.

# 4. Cadrage du Projet

## Objectif général

Développer un système de prédiction de liens capable d'identifier de nouvelles associations potentielles entre des gènes et des maladies.

## Données d'entrée

Le projet utilisera les données de la base HPO (Human Phenotype Ontology).

Fichier principal :

- genes_to_disease.txt

À partir de ce fichier, un graphe biparti gène-maladie sera construit.

## Tâche à réaliser

Pour chaque paire :

(gène, maladie)

non observée dans le graphe actuel, le système devra calculer un score représentant la probabilité qu'une association existe.

## Méthodes à comparer

### Méthode de référence (baseline)

Méthodes heuristiques simples :

- Common Neighbors
- Jaccard Coefficient

### Méthodes avancées

Méthodes basées sur l'apprentissage automatique :

- Node2Vec + Classifieur
- Graph Neural Network (optionnel)

## Sortie attendue

Pour une maladie donnée, le système devra retourner :

Top-K gènes candidats les plus probables

Exemple :

Maladie : Parkinson

Top 5 gènes candidats :

1. Gène A
2. Gène B
3. Gène C
4. Gène D
5. Gène E

## Critères d'évaluation

Les performances seront évaluées à l'aide de métriques classiques de Machine Learning :

- AUC-ROC
- Precision
- Recall
- F1-Score

Le projet sera considéré comme réussi si :

1. La méthode avancée dépasse la baseline heuristique.
2. Les gènes recommandés sont biologiquement cohérents.
3. Les résultats sont compatibles avec la littérature scientifique existante.

# Conclusion de la Phase 1

Cette première phase a permis :

- de comprendre le problème de prédiction de liens ;
- d'étudier son application aux associations gène-maladie ;
- d'explorer les principales approches utilisées dans la littérature ;
- de définir précisément le périmètre du projet.

La prochaine étape consistera à explorer les données HPO, construire le graphe gène-maladie et réaliser une analyse exploratoire du réseau.
